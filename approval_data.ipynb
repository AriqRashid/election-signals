```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Data/president.csv')
df.head()
```

```python
list(df)
```

```python
df[['start_date', 'end_date']]
```

```python
df= df[['end_date','sample_size','yes', 'no']].dropna()
df['end_date'] = pd.to_datetime(df['end_date'])
print(len(df))
df
```

```python
def weighted_avg(group, col):
    return (group[col] * group['sample_size']).sum() / group['sample_size'].sum()
```

```python
result = (
    df.groupby('end_date')
    .apply(lambda g: pd.Series({
        'yes': round(weighted_avg(g, 'yes'), 1),
        'no': round(weighted_avg(g, 'no'), 1)
    }), include_groups=False)
    .reset_index()
)
```

```python
result
```

```python
result.to_csv("Data/approvals_test_data.csv",  encoding='utf-8', index=False)
```
